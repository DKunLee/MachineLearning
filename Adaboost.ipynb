```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.ensemble import AdaBoostClassifier as sk_AdaBoostClassifier
from sklearn.linear_model import Perceptron as sk_Perceptron

from dklearn.ensemble import AdaBoostClassifier as dk_AdaBoostClassifier
from dklearn.linear_model import Perceptron as dk_Perceptron
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# AdaBoost from Scikit-Learn

```python
sk_estimator = sk_Perceptron(max_iter=10, eta0=0.1, random_state=42, tol=1e-3)

sk_ada = sk_AdaBoostClassifier(estimator=sk_estimator, n_estimators=50, learning_rate=1.0, algorithm='SAMME')

sk_ada.fit(X_train, y_train)
sk_y_pred = sk_ada.predict(X_test)

print("Accuracy:", accuracy_score(y_test, sk_y_pred))
```

# AdaBoost from Scratch

```python
dk_estimator = dk_Perceptron(max_iter=10, eta0=0.1, random_state=42)

dk_ada = dk_AdaBoostClassifier(estimator=dk_estimator, n_estimators=50)

dk_ada.fit(X_train, y_train)
dk_y_pred = dk_ada.predict(X_test)

print("Accuracy:", accuracy_score(y_test, dk_y_pred))
```
